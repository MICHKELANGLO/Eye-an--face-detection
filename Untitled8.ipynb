```python
#To use computer vision
import cv2
from google.colab.patches import cv2_imshow

cv2.__version__
```

```python
img_path= "/content/human pic.webp"        #To read image
img = cv2.imread(img_path)
face_cascade=cv2.CascadeClassifier('/haarcascade_frontalface_default.xml')
print("img shape:", img.shape)
gray=cv2.cvtColor(img, cv2.COLOR_RGB2GRAY) #convert to grayscale
cv2_imshow(img)
cv2_imshow(gray)
```

```python
faces=face_cascade.detectMultiScale(gray)
print(faces)
```

```python
for (x,y,w,h) in faces:
    cv2.rectangle(img,(x,y),(x+w,y+h),(250,100,0),2)
cv2_imshow(img)
```

```python
img1 = cv2.imread(img_path)
gray=cv2.cvtColor(img1, cv2.COLOR_RGB2GRAY)
eyes_cascade=cv2.CascadeClassifier('/haarcascade_eye.xml')
eyes =eyes_cascade.detectMultiScale(gray)
print(eyes)
for (x,y,w,h) in eyes:
    cv2.rectangle(img1,(x,y),(x+w,y+h),(250,100,0),2)
cv2_imshow(img1)
```

```python
def face_to_detect(img_path):
   img = cv2.imread(img_path)
   face_cascade=cv2.CascadeClassifier('/haarcascade_frontalface_default.xml')
   gray=cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
   faces=face_cascade.detectMultiScale(gray)
   for (x,y,w,h) in faces:
    cv2.rectangle(img,(x,y),(x+w,y+h),(250,100,0),2)
   cv2_imshow(img)
```

```python
def eyes_to_detect(img_path):
  img1 = cv2.imread(img_path)
  gray=cv2.cvtColor(img1, cv2.COLOR_RGB2GRAY)
  eyes_cascade=cv2.CascadeClassifier('/haarcascade_eye.xml')
  eyes =eyes_cascade.detectMultiScale(gray)
  for (x,y,w,h) in eyes:
    cv2.rectangle(img1,(x,y),(x+w,y+h),(250,100,0),2)
  cv2_imshow(img1)
```
